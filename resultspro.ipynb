```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Example ML Model Performance (Replace with actual values)
models = ["DT", "KNN", "LR", "SVM", "RFT","XG"]
accuracy = [0.9993,0.99907, 0.99912, 0.99837, 0.99940, 0.9995084442259752]  # Replace with actual accuracy values
f1_scores = [0.8057, 0.99907, 0.69512, 0.99649,0.82653,0.8444444444444444 ]  # Replace with actual F1-score values

# 1. Plot Accuracy & F1-Score Comparison (Like Figure 8)
plt.figure(figsize=(8, 6))
bar_width = 0.4
x_indexes = np.arange(len(models))

plt.bar(x_indexes, accuracy, width=bar_width, label="Accuracy Score", color='blue')
plt.bar(x_indexes + bar_width, f1_scores, width=bar_width, label="F1 Score", color='orange')

plt.xlabel("Models")
plt.ylabel("Accuracy and F1 Score")
plt.xticks(ticks=x_indexes + bar_width / 2, labels=models)
plt.legend()
plt.title("Accuracy Score of Machine Learning Models")
plt.show()



# 3. Confusion Matrix Visualization (Like Figure 6)
conf_matrices = {
    "Decision Tree": [[56836,28],[13,85]],
    "KNN": [[56758, 105], [0, 56863]],
    "Logistic Regression": [[56855, 9], [41, 57]],
    "SVM": [[17028, 34], [2, 5116]],
    "Random Forest": [[56847, 17], [17, 81]],
    "XGBoost":[[56858,6],[22,76]],
}

fig, axes = plt.subplots(2, 3, figsize=(12, 8))
axes = axes.flatten()

for i, (model, matrix) in enumerate(conf_matrices.items()):
    sns.heatmap(matrix, annot=True, fmt="d", cmap="Blues", ax=axes[i])
    axes[i].set_title(f"Confusion Matrix of {model}")
    axes[i].set_xlabel("Predicted Label")
    axes[i].set_ylabel("True Label")

plt.tight_layout()
plt.show()
```
